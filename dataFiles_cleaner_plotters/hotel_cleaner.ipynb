```python
hotelalldf = hoteldf.dropna()
print(hotelalldf.shape)
hotelalldf.sample(5)
```

```python
hotelalldf.isna().sum().sum() # cross check num of Nans
```

```python
#cleaning price, remove $ to use it as number
def remove_special_chars(val):
    new_val = re.findall('[0-9]',val)
    return "".join(new_val)

hotelalldf.price = hotelalldf.price.apply(remove_special_chars)
hotelalldf.head(5)
```

```python
#cleaning hotel star ,grab just numbers with decimal
def remove_chars(val):
    new_val = re.findall('[0-9.]+',val)
    return "".join(new_val)

hotelalldf.hotel_star = hotelalldf.hotel_star.apply(remove_chars)
hotelalldf.head(5)
```

```python
# grab numbers for num of ratings
hotelalldf.no_of_ratings = hotelalldf.no_of_ratings.apply(remove_special_chars)
hotelalldf.head(5)
```

```python
hotelalldf = hotelalldf.assign(rating = hotelalldf.rating.astype(float))
hotelalldf.head(3)
```

```python
hotelalldf.dtypes
```

```python
#Divide the hotel rating by 2, to have both airbnb and hotel ranked on a scale of 5
# must be run only once
hotelalldf = hotelalldf.assign(rating = hotelalldf.rating.div(2))
hotelalldf.head(3)
```

```python
def split_to_list(str_val):
    return str_val.split(",")
   
hotelalldf = hotelalldf.assign(occupancy_lst = hotelalldf.occupancy.apply(split_to_list))
```

```python
hotelalldf.head(5)
```

```python
# just initailzing new cols, assigned values are just place holders
hotelalldf = hotelalldf.assign(no_of_rooms = 1, no_of_guests =  3)
hotelalldf.sample(2)
```

```python
#convert no_of_reviews to int
hotelalldf.price=hotelalldf.price.astype(int)
```

```python
#convert hotel_star to float
hotelalldf.hotel_star=hotelalldf.hotel_star.astype(float)
```

```python
#convert no_of_ratings to int
hotelalldf.no_of_ratings=hotelalldf.no_of_ratings.astype(float)
```

```python
#convert no_of_rooms to float
hotelalldf.no_of_rooms=hotelalldf.no_of_rooms.astype(np.int32)
```

```python
#convert no_of_guests to float
hotelalldf.no_of_guests=hotelalldf.no_of_guests.astype(np.int32)
```
